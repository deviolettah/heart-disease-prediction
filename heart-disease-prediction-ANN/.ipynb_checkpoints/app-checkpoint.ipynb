```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn import model_selection
import os
```

```python
dataset = pd.read_csv('heart.csv', sep=';')
dataset.head()
```

```python
print( 'Shape of DataFrame: {}'.format(dataset.shape))
print (dataset.loc[1])
```

```python
# print 10 data terakhir
dataset.loc[295:]
```

```python
# remove missing data (indicated with a "?")
data = dataset[~dataset.isin(['?'])]
data.loc[295:]
```

```python
# drop rows with NaN values from DataFrame
data = data.dropna(axis=0)
data.loc[295:]
```

```python
# Ubah target 2,3,4 jadi 1
data.loc[data['target'] > 1, 'target'] = 1
data.loc[295:]
```

```python
# print the shape and data type of the dataframe
print(data.shape)
print(data.dtypes)
```

```python
# transform data to numeric to enable further analysis
data = data.apply(pd.to_numeric)
data.dtypes
```

```python
# print data characteristics, usings pandas built-in describe() function
data.describe()
```

```python
# plot histograms for each variable
data.hist(figsize = (12, 12))
plt.show()
```

```python
X = np.array(data.drop(['target'], 1))
y = np.array(data['target'])

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.2)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(16,input_dim=13, kernel_initializer='normal', activation='relu'),
  tf.keras.layers.Dense(32,kernel_initializer='normal', activation='relu'),
  tf.keras.layers.Dense(32,kernel_initializer='normal', activation='relu'),
  tf.keras.layers.Dense(8,kernel_initializer='normal', activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
adam = tf.keras.optimizers.Adam(lr=0.001)
model.compile(optimizer=adam,
                loss='binary_crossentropy',
                metrics=['acc'])
print(model.summary())
```

```python
# configure checkpoint untuk save file model
checkpoint_filepath = './Model_checkpoint'
checkpoint_prefix = os.path.join(checkpoint_filepath, "ckpt_{epoch}")
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_prefix,
    save_weights_only=True,
    monitor='val_acc',
    mode='max',
    save_best_only=True)
```

```python
history = model.fit(X_train, y_train,epochs=100,
                    validation_data=(X_test, y_test),
                    batch_size=10 ,
                    callbacks=[model_checkpoint_callback])
```

```python
%matplotlib inline
# Model accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```

```python
# Re-evaluate the model
loss,acc = model.evaluate(X,  y, verbose=2)
print("Restored model, accuracy: {:5.2f}%".format(100*acc))
```
